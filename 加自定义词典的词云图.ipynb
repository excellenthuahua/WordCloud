```python
from scipy.misc  import imread
import codecs
import jieba
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from jieba import analyse

stopwords = codecs.open(r'E:\text analyze\stopwords2.txt','r','utf-8').read()
jieba.load_userdict(r'E:\text analyze\自定义词典.txt')
text = codecs.open(r'E:\text analyze\凉茶类.txt','r','utf-8').read()

def process_word(text,stopwords):
    mywordlist = []
    seg_list = jieba.cut(text,cut_all = False)
    liststr = '/ '.join(seg_list) 
    for myword in liststr.split('/'):
        if not (myword.strip() in stopwords) and len(myword.strip()) > 1:
            mywordlist.append(myword)
    return "".join(mywordlist)

clean_word = process_word(text,stopwords)
tags = analyse.extract_tags(clean_word,topK = 30,withWeight=True,allowPOS=())
print(clean_word)
print(tags)

filename=r'E:\text analyze\易拉罐.jpg'
back_coloring=imread(filename)

#构建词云
wordcloud = WordCloud(font_path = r'E:\text analyze\msyh.ttc',
                   background_color = 'white',
                   width = 500,
                   height = 400,
                   mask = back_coloring,
                   margin = 2,
                   collocations=False).generate(clean_word)

plt.figure(figsize = (12,12))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
